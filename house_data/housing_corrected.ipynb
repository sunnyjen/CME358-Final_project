```python
import pandas as pd

houses = pd.read_csv('geocoded_addresses_housesigma.csv')
display(houses)
houses.reset_index(inplace=True)
# Rename columns to generic names like Column_0, Column_1, ...
houses.columns = [f"Column_{i}" for i in range(len(houses.columns))]

# Rename specific columns based on your intended structure
# Adjust the names in the dictionary based on your CSV structure
houses.rename(columns={
    'Column_0': 'link',
    'Column_1': 'Listed Price',
    'Column_2': 'Sold Price',
    'Column_3': 'Type',
    'Column_4': 'address',
    'Column_5': 'Lat',
    'Column_6': 'Long'
}, inplace=True)

# Display the renamed DataFrame
display(houses)
houses.to_csv('housesigma_data_with_coords.csv')
```
