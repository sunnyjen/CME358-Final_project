```python
import pdfplumber
import pandas as pd
from pathlib import Path
#global variables
```

```python
current_dir = Path.cwd()
```

```python

# Get the current working directory
# Get the parent directory
parent_dir = current_dir.parent
csv_folder = parent_dir/'data'/'csv'
pdf_path = parent_dir / "data" / "pdf" / "2021_EQAO.pdf"

all_tables = []
 
# Open the PDF with pdfplumber
with pdfplumber.open(pdf_path) as pdf:
    for page in pdf.pages:
        # Extract tables from the current page
        tables = page.extract_tables()
       
        # If tables are found, add them to the list
        for table in tables:
            all_tables.append(table)

# Convert the first table to a pandas DataFrame (assuming the first table is your target)
df = pd.DataFrame(all_tables[0][1:], columns=all_tables[0][0])  # Use the first row as headers
df.to_csv(csv_folder/'2021_EQAO.csv', index=False)  # index=False excludes the DataFrame index from the CSV file

   
# Display the DataFrame
df
```
